```python
import pandas as pd
from glob import glob
```

```python
tax_csv = glob('fastgather/*.tax.csv')

df_list = []
for filename in tax_csv:
    df_list.append(pd.read_csv(filename))

tax_df = pd.concat(df_list)
```

```python
tax_df = tax_df[tax_df.status == 'match']
tax_df
```

```python
def get_species(lineage_str):
    species = lineage_str.split(';')[-1]
    assert species.startswith('s__')
    return species

def get_genus(lineage_str):
    genus = lineage_str.split(';')[-2]
    assert genus.startswith('g__')
    return genus
```

```python
tax_df['species'] = tax_df['lineage'].apply(get_species)
tax_df['genus'] = tax_df['lineage'].apply(get_genus)
```

```python
tax_df
```

```python
species_df = tax_df.groupby(['species'])['species'].count()
```
